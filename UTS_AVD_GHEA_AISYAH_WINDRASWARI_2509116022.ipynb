# **Data Preparation**

### **Inconsistent Values**

Pada bagian ini, dilakukan penanganan terhadap beberapa kolom pada dataset kesehatan mental mahasiswa:

* `Gender`, karena terdapat kemungkinan penulisan gender yang tidak konsisten seperti `Male`, `male`, `M`, atau `Female`, `female`. Maka dari itu dilakukan penanganan dengan menyamakan seluruh penulisan menjadi format baku seperti `Male` dan `Female` agar memudahkan proses analisis.

* `Course`, karena terdapat beberapa nama jurusan yang ditulis dengan format berbeda atau singkatan yang tidak konsisten. Maka dari itu dilakukan penanganan dengan merapikan nama jurusan menjadi format yang seragam agar tidak terjadi duplikasi kategori saat visualisasi.

* `Year`, karena format tahun studi tidak konsisten seperti `Year 1`, `year 1`, atau `Year1`. Maka dari itu dilakukan penanganan dengan menyamakan seluruh format penulisan agar lebih rapi dan mudah dianalisis.

* `CGPA`, karena nilai IPK masih berbentuk rentang teks seperti `3.50 - 4.00` sehingga belum dapat digunakan untuk analisis numerik. Maka dari itu dilakukan penanganan dengan membersihkan format penulisan dan mempersiapkan data agar dapat dikonversi menjadi numerik.

* `Depression`, `Anxiety`, dan `Panic_Attack`, karena data masih berbentuk teks `Yes/No` dengan kemungkinan perbedaan huruf besar kecil. Maka dari itu dilakukan penanganan dengan menyeragamkan format penulisan agar dapat dilakukan encoding dan analisis lebih lanjut.

* `Marital_Status`, karena terdapat kemungkinan variasi penulisan status pernikahan seperti `Yes`, `No`, atau penulisan yang tidak konsisten. Maka dari itu dilakukan penanganan dengan menyederhanakan nilai menjadi format kategorikal yang seragam.

* `Timestamp`, masalahnya adalah data masih berbentuk teks (*object*) sehingga belum dapat digunakan untuk analisis waktu. Maka dari itu dilakukan penanganan dengan mengubah format menjadi tipe data `datetime` agar dapat dianalisis berdasarkan waktu pengisian survei.


```python
# Rapikan isi kolom Gender
df['Gender'] = df['Gender'].str.strip().str.title()

# Samakan penulisan gender
gender_mapping = {
    'Male': 'Male',
    'M': 'Male',
    'Female': 'Female',
    'F': 'Female'
}

df['Gender'] = df['Gender'].replace(gender_mapping)

# Cek hasil
df['Gender'].unique()
```

```python
# Rapikan format Year
df['Year'] = df['Year'].astype(str).str.lower().str.strip()

year_mapping = {
    'year 1': 'Year 1',
    'year 2': 'Year 2',
    'year 3': 'Year 3',
    'year 4': 'Year 4',
    '1': 'Year 1',
    '2': 'Year 2',
    '3': 'Year 3',
    '4': 'Year 4'
}

df['Year'] = df['Year'].replace(year_mapping)

# Cek hasil
df['Year'].unique()
```

```python
# Rapikan format CGPA
df['CGPA'] = df['CGPA'].astype(str).str.strip()

# Samakan separator
df['CGPA'] = df['CGPA'].str.replace('–', '-', regex=False)
df['CGPA'] = df['CGPA'].str.replace('—', '-', regex=False)

# Cek hasil
df['CGPA'].unique()
```

```python
mental_cols = ['Depression', 'Anxiety', 'Panic_Attack', 'Treatment']

for col in mental_cols:
    df[col] = df[col].astype(str).str.strip().str.title()

# Mapping Yes/No
yes_no_mapping = {
    'Yes': 'Yes',
    'Y': 'Yes',
    'No': 'No',
    'N': 'No'
}

for col in mental_cols:
    df[col] = df[col].replace(yes_no_mapping)

# Cek hasil
df[mental_cols].head()
```

```python
# Rapikan nama jurusan
df['Course'] = df['Course'].astype(str).str.upper().str.strip()

# Cek hasil
df['Course'].unique()
```

```python
# Ubah Timestamp menjadi datetime
df['Timestamp'] = pd.to_datetime(
    df['Timestamp'],
    errors='coerce'
)

# Cek hasil
df.info()
```

### **Outliers Values**

Pada tahap pengecekan *outliers*, dilakukan analisis pada kolom numerik `Age` untuk mengetahui apakah terdapat data usia yang memiliki nilai terlalu jauh dari mayoritas data lainnya.

Berdasarkan hasil pengecekan, persentase outlier pada kolom `Age` adalah **0.0%**, yang berarti tidak ditemukan data usia yang menyimpang secara ekstrem. Seluruh data usia masih berada dalam rentang yang wajar untuk kategori mahasiswa, yaitu sekitar 18–24 tahun.

Hal ini menunjukkan bahwa distribusi data usia pada dataset cukup stabil dan tidak memerlukan penanganan khusus seperti penghapusan atau transformasi data outlier. Dengan demikian, data usia dapat langsung digunakan untuk proses analisis dan visualisasi tanpa khawatir memengaruhi hasil secara signifikan.


```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```
